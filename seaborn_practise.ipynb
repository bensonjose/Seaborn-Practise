# What is the seaborn used for?

Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures. Seaborn helps you explore and understand your data.




# Why use seaborn and matplotlib?
Matplotlib and Seaborn are two of the most widely used visualization libraries in Python. They both allow you to quickly perform data visualization for gaining statistical insights and telling a story with data.

# What is the advantage of seaborn?
Advantages of seaborn

**It has a concise and expressive syntax that allows users to create complex and customized plots quickly. Additionally, seaborn integrates seamlessly with pandas, the main data manipulation library in Python, and can use pandas data structures as input.**

# What does KDE mean in Seaborn?
kernel density estimate
A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.

***KDE represents the data using a continuous probability density curve in one or more dimensions.***

```python
 sns.histplot(tips['tip'],kde=True,bins=20)
```

# Jointplot

# The Seaborn.jointplot() method
**is used to subplot grid for plotting pairwise relationships in a dataset. This function offers the JointGrid class a handy interface with a number of pre-made plot types. If you require greater flexibility, you should use JointGrid directly as this wrapper is designed to be relatively lightweight.**

As the name suggests, the Joint Plot is comprised of three individual plots â€“ a bivariate graph that shows the difference of the dependent variableâ€™s behavior with independent variableâ€™s behavior, another plot placed above the bivariate graph to observe the distribution of independent variables, another plot positioned on the right margin vertically that displays the distribution of the dependent variable.

```python
sns.jointplot(x="tip",y="total_bill",data=tips)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade="YlGnBu",cmap="YlGnBu")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",cmap="YlGnBu")
```
